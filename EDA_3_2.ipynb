```python
df = pd.read_csv("train.csv")
```

```python
print("Mediana ceny:", df["SalePrice"].median())
```

```python
# Typy danych wszystkich kolumn
df.info()
```

```python
# Histogram SalePrice
for b in [10, 30, 50]:
    plt.hist(df["SalePrice"], bins=b, color='skyblue', edgecolor='black')
    plt.title(f"Rozkład cen domów (bins={b})")
    plt.xlabel("Cena")
    plt.ylabel("Liczba domów")
    plt.show()
```

```python
print("Skośność SalePrice:", df["SalePrice"].skew())
```

```python
# Boxplot SalePrice
sns.boxplot(x=df["SalePrice"])
plt.title("SalePrice – Boxplot")
plt.show()
```

**Analiza**

1. **Rozkład jest prawoskośny** - większość domów jest tańszych a tylko kilka bardzo drogich. Widać to wyraźnie dla każdej liczby przedziałów.
2. **Średnia cena** jest wyższa od **mediany**, widać więc wpływ tych kilku drogich domów na średnią.
3. Na boxplocie widać kilka wartości odstających - są to te bardzo drogie domy, które znajdują się daleko od reszty danych.
4. Typowy **zakres cen (między Q1 a Q3)** wynosi około 125 000 - 220 000.
5. **Rozkład nie jest normalny** i **symetryczny**
6. Boxplot jest poprawny, większość domów jest przeciętnych, ale istnieją też takie bardzo drogie.

```python
# Scatter GrLivArea vs SalePrice
sns.scatterplot(x="GrLivArea", y="SalePrice", data=df)
sns.regplot(x="GrLivArea", y="SalePrice", data=df, line_kws={"color": "red"})
plt.title("Zależność między powierzchnią mieszkalną a ceną + linia trendu")
plt.xlabel('Powierzchnia mieszkalna [sq ft]')
plt.ylabel('Cena [USD]')
plt.show()
```

```python
corr_value = df["GrLivArea"].corr(df["SalePrice"])
print(f"Korelacja GrLivArea <-> SalePrice: {corr_value:.3f}")
```

```python
df[(df["GrLivArea"] > 4000) & (df["SalePrice"] < 300000)][["GrLivArea", "SalePrice"]]
```

**Interpretacja**

1. Istnieje wyraźna zależność nie tylko wizualna, ale również oparta na oblieczeniach. Jak wynika z wartości korelacji mamy silną dodatnią zależność, mówiąca że **im większy dom, tym cena jest wyższa**. Jak można zaobserwowac po linii trendu większośc punktów układa się wzdłuż tej proste, więc zależnośc** ma charakter liniowy**.
2. Jak widać na scatterze mamy dwa dziwne elementy, które pomimo dużej przestrzeni mieszkalnej mają niską cenę. Może to świadczyć o tym że są np. zniszczone czy niewykończone.

```python
# Scatter LotArea vs SalePrice 1
sns.scatterplot(x="LotArea", y="SalePrice", data=df)
sns.regplot(x="LotArea", y="SalePrice", data=df, line_kws={"color": "red"})
plt.title("Zależność między powierzchnią a ceną + linia trendu (skala liniowa)")
plt.xlabel('Powierzchnia działki [sq ft]')
plt.ylabel('Cena [USD]')
plt.show()
```

```python
# Scatter LotArea vs SalePrice 2
sns.scatterplot(x="LotArea", y="SalePrice", data=df)
sns.regplot(x="LotArea", y="SalePrice", data=df, line_kws={"color": "red"})
plt.xscale("log")
plt.title("Zależność między powierzchnią działki a ceną (skala logarytmiczna)")
plt.xlabel('Powierzchnia działki [sq ft]')
plt.ylabel('Cena [USD]')
plt.show()
```

```python
corr_value = df["LotArea"].corr(df["SalePrice"])
print(f"Korelacja LotArea <-> SalePrice: {corr_value:.3f}")
```

**Interpetacje dodatkowe 1**
1. Wykres **LotArea vs SalePrice** w skali liniowej wygląda dość nieczytelnie, ponieważ kilka bardzo dużych działek zaburza skalę, lecz po zastosowaniu skali logarytmicznej widac, że większe działki mają tendencję do wyższych cen ale związek jest słaby, co widać również po współczynniku korelacji.

```python
fig, axes = plt.subplots(1, 3, figsize=(16,5))

sns.scatterplot(ax=axes[0], x="TotalBsmtSF", y="SalePrice", data=df, alpha=0.6, color="orange")
sns.regplot(ax=axes[0], x="TotalBsmtSF", y="SalePrice", data=df, scatter=False, line_kws={"color": "red"})
axes[0].set_title("TotalBsmtSF vs SalePrice")
axes[0].set_xlabel("Powierzchnia piwnicy [sq ft]")
axes[0].set_ylabel("Cena [USD]")

sns.scatterplot(ax=axes[1], x="1stFlrSF", y="SalePrice", data=df, alpha=0.6, color="royalblue")
sns.regplot(ax=axes[1], x="1stFlrSF", y="SalePrice", data=df, scatter=False, line_kws={"color": "red"})
axes[1].set_title("1stFlrSF vs SalePrice")
axes[1].set_xlabel("Powierzchnia 1. piętra [sq ft]")
axes[1].set_ylabel("Cena [USD]")

sns.scatterplot(ax=axes[2], x="GarageArea", y="SalePrice", data=df, alpha=0.6, color="green")
sns.regplot(ax=axes[2], x="GarageArea", y="SalePrice", data=df, scatter=False, line_kws={"color": "red"})
axes[2].set_title("GarageArea vs SalePrice")
axes[2].set_xlabel("Powierzchnia garażu [sq ft]")
axes[2].set_ylabel("Cena [USD]")

plt.tight_layout()
plt.show()
```

```python
cols = ["TotalBsmtSF", "1stFlrSF", "GarageArea", "SalePrice"]
corrs = df[cols].corr()["SalePrice"].sort_values(ascending=False)
print(corrs)
```

**Interpetacje dodatkowe 2**
1. Wszystkie powierzchnie są dodatnio skorelowane z ceną - co ma sens, bo zazwyczaj większa powierzchnia oznacza większa cene.
2. Najsilniejszy związek ma **GarageArea** , ale kolejne **TotalBsmtSF** i **1stFlrSF** są prawie równie mocno związane.
3. Specyficznym elementem są pionowe punkty przy 0 powierzchni garażu oraz piwnicy, ale są one zrozumiałe i dotyczą po prostu sytuacji gdy nie ma garażu czy piwnicy.
4. Zależności są prawie liniowe, lecz widać kilka odstających punktów w przypadku każdego z wykresów.
