```python
# Histogram SalePrice
for b in [10, 30, 50]:
    plt.hist(df["SalePrice"], bins=b, color='skyblue', edgecolor='black')
    plt.title(f"Rozkad cen domw (bins={b})")
    plt.xlabel("Cena")
    plt.ylabel("Liczba domw")
    plt.show()
```

```python
print("Skono SalePrice:", df["SalePrice"].skew())
```

```python
# Boxplot SalePrice
sns.boxplot(x=df["SalePrice"])
plt.title("SalePrice  Boxplot")
plt.show()
```

**Analiza**

1. **Rozkad jest prawoskony** - wikszo domw jest taszych a tylko kilka bardzo drogich. Wida to wyranie dla kadej liczby przedziaw.
2. **rednia cena** jest wysza od **mediany**, wida wic wpyw tych kilku drogich domw na redni.
3. Na boxplocie wida kilka wartoci odstajcych - s to te bardzo drogie domy, ktre znajduj si daleko od reszty danych.
4. Typowy **zakres cen (midzy Q1 a Q3)** wynosi okoo 125 000 - 220 000.
5. **Rozkad nie jest normalny** i **symetryczny**
6. Boxplot jest poprawny, wikszo domw jest przecitnych, ale istniej te takie bardzo drogie.

```python
# Scatter GrLivArea vs SalePrice
sns.scatterplot(x="GrLivArea", y="SalePrice", data=df)
sns.regplot(x="GrLivArea", y="SalePrice", data=df, line_kws={"color": "red"})
plt.title("Zaleno midzy powierzchni mieszkaln a cen + linia trendu")
plt.xlabel('Powierzchnia mieszkalna [sq ft]')
plt.ylabel('Cena [USD]')
plt.show()
```

```python
corr_value = df["GrLivArea"].corr(df["SalePrice"])
print(f"Korelacja GrLivArea <-> SalePrice: {corr_value:.3f}")
```

**Interpretacja**

1. Istnieje wyrana zaleno nie tylko wizualna, ale rwnie oparta na oblieczeniach. Jak wynika z wartoci korelacji mamy siln dodatni zaleno, mwica e **im wikszy dom, tym cena jest wysza**. Jak mona zaobserwowac po linii trendu wikszoc punktw ukada si wzdu tej proste, wic zalenoc** ma charakter liniowy**.
2. Jak wida na scatterze mamy dwa dziwne elementy, ktre pomimo duej przestrzeni mieszkalnej maj nisk cen. Moe to wiadczy o tym e s np. zniszczone czy niewykoczone.

```python
# Scatter LotArea vs SalePrice 1
sns.scatterplot(x="LotArea", y="SalePrice", data=df)
sns.regplot(x="LotArea", y="SalePrice", data=df, line_kws={"color": "red"})
plt.title("Zaleno midzy powierzchni a cen + linia trendu (skala liniowa)")
plt.xlabel('Powierzchnia dziaki [sq ft]')
plt.ylabel('Cena [USD]')
plt.show()
```

```python
# Scatter LotArea vs SalePrice 2
sns.scatterplot(x="LotArea", y="SalePrice", data=df)
sns.regplot(x="LotArea", y="SalePrice", data=df, line_kws={"color": "red"})
plt.xscale("log")
plt.title("Zaleno midzy powierzchni dziaki a cen (skala logarytmiczna)")
plt.xlabel('Powierzchnia dziaki [sq ft]')
plt.ylabel('Cena [USD]')
plt.show()
```

```python
corr_value = df["LotArea"].corr(df["SalePrice"])
print(f"Korelacja LotArea <-> SalePrice: {corr_value:.3f}")
```

**Interpetacje dodatkowe 1**
1. Wykres **LotArea vs SalePrice** w skali liniowej wyglda do nieczytelnie, poniewa kilka bardzo duych dziaek zaburza skal, lecz po zastosowaniu skali logarytmicznej widac, e wiksze dziaki maj tendencj do wyszych cen ale zwizek jest saby, co wida rwnie po wspczynniku korelacji.

```python
fig, axes = plt.subplots(1, 3, figsize=(16,5))

sns.scatterplot(ax=axes[0], x="TotalBsmtSF", y="SalePrice", data=df, alpha=0.6, color="orange")
sns.regplot(ax=axes[0], x="TotalBsmtSF", y="SalePrice", data=df, scatter=False, line_kws={"color": "red"})
axes[0].set_title("TotalBsmtSF vs SalePrice")
axes[0].set_xlabel("Powierzchnia piwnicy [sq ft]")
axes[0].set_ylabel("Cena [USD]")

sns.scatterplot(ax=axes[1], x="1stFlrSF", y="SalePrice", data=df, alpha=0.6, color="royalblue")
sns.regplot(ax=axes[1], x="1stFlrSF", y="SalePrice", data=df, scatter=False, line_kws={"color": "red"})
axes[1].set_title("1stFlrSF vs SalePrice")
axes[1].set_xlabel("Powierzchnia 1. pitra [sq ft]")
axes[1].set_ylabel("Cena [USD]")

sns.scatterplot(ax=axes[2], x="GarageArea", y="SalePrice", data=df, alpha=0.6, color="green")
sns.regplot(ax=axes[2], x="GarageArea", y="SalePrice", data=df, scatter=False, line_kws={"color": "red"})
axes[2].set_title("GarageArea vs SalePrice")
axes[2].set_xlabel("Powierzchnia garau [sq ft]")
axes[2].set_ylabel("Cena [USD]")

plt.tight_layout()
plt.show()
```

```python
cols = ["TotalBsmtSF", "1stFlrSF", "GarageArea", "SalePrice"]
corrs = df[cols].corr()["SalePrice"].sort_values(ascending=False)
print(corrs)
```

**Interpetacje dodatkowe 2**
1. Wszystkie powierzchnie s dodatnio skorelowane z cen - co ma sens, bo zazwyczaj wiksza powierzchnia oznacza wiksza cene.
2. Najsilniejszy zwizek ma **GarageArea** , ale kolejne **TotalBsmtSF** i **1stFlrSF** s prawie rwnie mocno zwizane.
3. Specyficznym elementem s pionowe punkty przy 0 powierzchni garau oraz piwnicy, ale s one zrozumiae i dotycz po prostu sytuacji gdy nie ma garau czy piwnicy.
4. Zalenoci s prawie liniowe, lecz wida kilka odstajcych punktw w przypadku kadego z wykresw.
